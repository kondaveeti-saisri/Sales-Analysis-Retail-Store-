```python
df = pd.read_csv(r"V:\sai\dsprojects\retailsales\Warehouse_and_Retail_Sales.csv")
```

```python
print(df.head())
```

```python
print(df.isnull().sum())

```

```python
df['SUPPLIER'] = df['SUPPLIER'].fillna("Unknown")
df['ITEM TYPE'] = df['ITEM TYPE'].fillna(df['ITEM TYPE'].mode()[0])
df['RETAIL SALES'] = df['RETAIL SALES'].fillna(0)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
monthly_sales = df.groupby('MONTH')['RETAIL SALES'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

# Plot monthly sales
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Retail Sales")
plt.title("Monthly Retail Sales Trend")
plt.xticks(range(1, 13))  # Ensure x-axis shows all months from 1 to 12
plt.grid(True)

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Group sales by ITEM TYPE
category_sales = df.groupby('ITEM TYPE')['RETAIL SALES'].sum().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette="viridis")

# Add labels and title
plt.xlabel("Item Type")
plt.ylabel("Retail Sales")
plt.title("Total Sales by Item Type")
plt.xticks(rotation=45)  # Rotate for better readability

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group sales by ITEM TYPE
category_sales = df.groupby('ITEM TYPE')['RETAIL SALES'].sum().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x=category_sales.index, y=category_sales.values, hue=category_sales.index, palette="viridis", legend=False)

# Add labels and title
plt.xlabel("Item Type")
plt.ylabel("Retail Sales")
plt.title("Total Sales by Item Type")
plt.xticks(rotation=45)  # Rotate for better readability

# Show the plot
plt.show()
```

```python
# Group sales by SUPPLIER and sort by highest sales
supplier_sales = df.groupby('SUPPLIER')['RETAIL SALES'].sum().sort_values(ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=supplier_sales.values, y=supplier_sales.index, palette="coolwarm")

# Add labels and title
plt.xlabel("Retail Sales")
plt.ylabel("Supplier")
plt.title("Top 10 Suppliers by Retail Sales")

# Show the plot
plt.show()
```
